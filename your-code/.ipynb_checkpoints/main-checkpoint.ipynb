To perform time series analysis on the data, we must change the index from a range index to a time series index. In the cell below, create a time series index using the `pd.date_range` function. Create a time series index starting at 1/1/2018 00:00:00 and ending at 1/1/2018 00:29:10. The number of periods is equal to the number of rows in `sensor`. The frequency should be set to `infer`.

```python
# Creating a pandas series for the index with a range of dates and time:

DATE = pd.date_range(start='1/1/2018 00:00:00', end='1/1/2018 00:29:10', periods=1751)
```

Assign the time series index to the dataframe's index.

```python
#setting the index as the first column 
sensor.index = pd.DatetimeIndex(DATE, freq='infer')
```

Our next step is to decompose the time series and evaluate the patterns in the data. Load the `statsmodels.api` submodule and plot the decomposed plot of `userAcceleration.x`. Set `freq=60` in the `seasonal_decompose` function. Your graph should look like the one below.

![time series decomposition](../images/tsa_decompose.png)

```python
sensor.head(1)
```

```python
ts = sensor['userAcceleration.x']
```

```python
# Your code here:

from statsmodels.tsa.seasonal import seasonal_decompose
decomposition = seasonal_decompose(ts, period=60)

trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid
```

```python
import matplotlib.pyplot as plt

from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 15, 6

#plotting every import plot to see if there is seasonality

plt.figure(figsize=(15,10))
plt.subplot(411)
#plot1
plt.plot(ts, label='Original') 
plt.legend(loc='best')
plt.subplot(412)

#plot2
plt.plot(trend, label='Trend')
plt.legend(loc='best')
plt.subplot(413)

#plot3
plt.plot(seasonal,label='Seasonality')
plt.legend(loc='best')
plt.subplot(414)

#plot4
plt.plot(residual, label='Residuals')
plt.legend(loc='best')
plt.tight_layout()
```

Plot the decomposed time series of `rotationRate.x` also with a frequency of 60.

```python
from statsmodels.tsa.seasonal import seasonal_decompose
decomposition = seasonal_decompose(sensor['rotationRate.x'], period=60)

trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid
```

```python
import matplotlib.pyplot as plt

from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 15, 6

#plotting every import plot to see if there is seasonality

plt.figure(figsize=(15,10))
plt.subplot(411)
#plot1
plt.plot(sensor['rotationRate.x'], label='Original') 
plt.legend(loc='best')
plt.subplot(412)

#plot2
plt.plot(trend, label='Trend')
plt.legend(loc='best')
plt.subplot(413)

#plot3
plt.plot(seasonal,label='Seasonality')
plt.legend(loc='best')
plt.subplot(414)

#plot4
plt.plot(residual, label='Residuals')
plt.legend(loc='best')
plt.tight_layout()
```

The next step will be to test both variables for stationarity. Perform the Augmented Dickey Fuller test on both variables below.

```python
# Your code here:
from statsmodels.tsa.stattools import adfuller
```

```python
adfuller_userAcceleration = adfuller(sensor['userAcceleration.x'])
print("adfuller_userAcceleration:",adfuller_userAcceleration, "\n")

adfuller_rotationRate = adfuller(sensor['rotationRate.x'])
print("adfuller_rotationRate:", adfuller_rotationRate)
```

What are your conclusions from this test?

```python
# Your conclusions here:

'''
We can reject the fact that the seasonality is not stationary, so we are good to perform a prediction of the 
future values since we have stationary values based in a times series. 

This is because both p-values are less than the critical value.
'''
```
